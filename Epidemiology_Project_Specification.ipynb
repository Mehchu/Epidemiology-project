```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.optimize import newton
from scipy.optimize import fsolve
```

```python
# --- Problem setup --- #

# set up time grid for the solution
tf = 100
Nsteps = 1000
t = np.linspace(0, tf, Nsteps+1)

# Set initial conditions and put into array y0
beta = 2*(10**-5)
mu = 0.1

S0 = 10**4 - 3
I0 = 3
R0 = 0
y0 = np.array([S0, I0, R0])
```

```python
# --- Define the RHS of ODE system --- #

def SIR(y, t):
    # returns RHS of the SIR model
    S, I, R = y

    N = S + I + R
    Sdot = -beta * S * I
    Idot = beta * S * I - mu * I
    Rdot = mu * I
    
    return Sdot, Idot, Rdot
```

```python
# --- Call the ODE solver and extract solutions --- #
  
y_sol = odeint(SIR, y0, t)

# extract S, I, R from the solution array y
S = y_sol[:,0]
I = y_sol[:,1]
R = y_sol[:,2]
```

```python
# --- Plot data --- #

# plot the time evolution of the S, I, R
plt.plot(t, S, label = "S")
plt.plot(t, I, lw = 4, label = "I")
plt.plot(t, R, label = "R")

# labels etc
plt.xlabel("time", fontsize = 14)
plt.ylabel("S,I,R", fontsize = 14)
plt.title("SIR model", fontsize = 16)
plt.legend(fontsize = 14)

# Extract the maximum I value
indexMaxI = I.argmax()
# Divide by 10 to scale the time axis appropriately and plot the maximum point
plt.plot(indexMaxI/10, I[indexMaxI], 'k', marker = 'o', markersize = 10)

plt.show()        

# plot the evolution in the S,I phase plane
plt.plot(S, I)
# plot the maximum point in the S,I phase plane
plt.plot(S[indexMaxI], I[indexMaxI], 'k', marker = 'o', markersize = 10)

# labels 
plt.xlabel("S", fontsize = 14)
plt.ylabel("I", fontsize = 14)
plt.title("Phase plane for SIR model", fontsize = 16)

plt.show()

# Output the maximum I value rounded down
print(f"The maximum I value is {int(np.floor(I[indexMaxI]))}")
```

```python
# Create a range of beta values
betas = np.linspace(0, 1*(10**-3), 5000+1)

# Appropriately label the plot
plt.xlabel("beta", fontsize = 14)
plt.ylabel("max I", fontsize = 14)
plt.title("I-beta dependence", fontsize = 16)

# Initialise an array to store the maximum I value
maxIs = np.array([])

for beta in betas:
    
    # Solve the ODE for the current beta
    y_sol = odeint(SIR, y0, t)

    # Extract the I values from the solution
    I = y_sol[:,1]
    
    # Extract the maximum I value
    maxI = I.max()
    
    # Store the maximum I value in an array
    maxIs = np.append(maxIs, maxI)
    
# Plot the maximum value of I against the corresponding beta value
plt.plot(betas, maxIs)

plt.show()
```

## Derivation for Final Size Equation with V compartment

Starting with the known derivatives:
$$ \frac{dS}{dt}=-\beta IS $$
$$ \frac{dI}{dt}=\beta IS-\mu I $$
We can derive $\frac{dI}{dS}=-1+\frac\mu{\beta S}$ which integrating gives,
$$I=-S+{\frac\mu\beta}\ln(S)+C$$
With inital conditions $C=I(0)+S(0)-{\frac\mu\beta}\ln(S(0))$.

Substituting in for C we get,
$$I=-S+{\frac\mu\beta}\ln(S)+I(0)+S(0)-{\frac\mu\beta}\ln(S(0))$$

At the start of the outbreak, we have $I(0)+S(0)=N-V$, so we can write,
$$I=-S+{\frac\mu\beta}\ln(S)+ (N-V)-{\frac\mu\beta}\ln(S(0))$$
As $t\rightarrow\infty$ we have $I(t)\rightarrow0$ and $S(t)+R(t)\rightarrow(N-V)$. So taking the limit as $t\rightarrow\infty$,
$$\lim\limits_{t\rightarrow\infty}[-((N-V)-R(t))+{\frac\mu\beta}\ln((N-V)-R(t))+(N-V)-{\frac\mu\beta}\ln(S(0))]=0$$
Rearranging yields:
$$\lim\limits_{t\rightarrow\infty}R(t)=(N-V)-S(0)exp\Biggl(\frac{-\beta\lim\limits_{t\rightarrow\infty}R(t)}\mu\Biggl)$$
Letting $\lim\limits_{t\rightarrow\infty}R(t)=R'$ where $R'$ is the Final Size, we can write,
$$R'=(N-V)-S(0)exp\biggl(\frac{-\beta R'}\mu\biggl)$$
Approximating $S(0)\approx (N-V)$ yields us our Final Size Equation of:
$$R'=(N-V)-(N-V)exp\biggl(\frac{-\beta R'}\mu\biggl)$$

```python
mu = 0.1
N = 10**6
S0, I0, R0 = N - 5, 5, 0 
y0 = [S0, I0, R0]
tf = 300
Nsteps = 1000
t = np.linspace(0, tf, Nsteps+1)

def beta(t, T):
    if t < T:
        return 2*(10**-7)
    elif T <= t <= T + 30:
        return 1.5*(10**-7)
    else:
        return 2*(10**-7)

def SIR_varbeta(y, t, T):
    S, I, R = y
    beta_t = beta(t, T)
    Sdot = -beta_t * S * I
    Idot = beta_t * S * I - mu * I
    Rdot = mu * I
    return [Sdot, Idot, Rdot]

# Run the model for T = 30
T_fixed = 30
y_sol = odeint(SIR_varbeta, y0, t, args=(T_fixed,))
S, I, R = y_sol[:, 0], y_sol[:, 1], y_sol[:, 2]

plt.plot(t, S, label="S")
plt.plot(t, I, label="I")
plt.plot(t, R, label="R")
plt.xlabel("Time (days)")
plt.ylabel("S,I,R")
plt.title("SIR Model with Social Distancing (T = 30 days)")
plt.legend()
plt.show()

# Find the value of T which minimize maximum value of I
T_values = np.arange(1, 201)
max_I_values = []

for T in T_values:
    y_sol = odeint(SIR_varbeta, y0, t, args=(T,))
    max_I_values.append(max(y_sol[:, 1]))

value_of_T = T_values[np.argmin(max_I_values)]
print(f"The value of T that flattens the curve most: {value_of_T}")

plt.plot(value_of_T, min(max_I_values), 'ro', markersize = 5)
plt.plot(T_values, max_I_values,color='b')
plt.xlabel("T (days)")
plt.ylabel("Maximum Infected Population")
plt.title("Effect of Social Distancing on Peak Infections in variable T")
plt.show()
```

```python
# --- Stochastic SIS model (Bullet point 3) --- #
# !!WARNING!! Code will take a couple of hours to run

# Setting up constants
N = 1000
mu = 0.1
I = 1
Sim = 10000 #No. of simulations
R0_ = [0.8, 1.01, 1.1, 1.3, 1.6, 2]
D = 6000 #Max No. of days
t_ = range(1,D+1)

# Function for finding betas
def findbeta(R):
    beta = R/(N*(1/mu))
    return beta

betas_ = [findbeta(r0) for r0 in R0_]

# Setting up simulations for SIS model to record duration in an array
T = []
for B in betas_:
    TotalDays = np.array([])

    for i in range(Sim):
        S = N - 1
        I = 1
        t = 0

        while I > 0 and t < D:
            r1 = np.random.uniform(0, 1, 1)
            totalRate = B*I*S + mu*I
            t = t - (1/totalRate)*np.log(r1)

            r2 = np.random.uniform(0, 1, 1)
            if r2 < B*I*S/totalRate:
                S = S - 1
                I = I + 1
            else:
                I = I - 1
                S = S + 1

        TotalDays = np.append(TotalDays, t)

    T.append(TotalDays)

# Function to calculate probability of at least t days
def t_probability(beta_index, t):
    times = T[beta_index]
    counter = 0

    for i in times:
        if i >= t:
            counter += 1

    return counter / Sim

# Function to create array of t_probability for all t
def prob_array(beta_index):
    Prob=[]
    
    for t in range(1,D+1):
        Prob.append(t_probability(beta_index, t))
        
    return Prob

# Array containing all data of t probability
prob_=[prob_array(n) for n in range(len(betas_))]

# Plotting of the graph
for i in range(len(betas_)):
    plt.plot(t_, prob_[i], label = "R0="+str(R0_[i]))
    
plt.title("Probability of a severe epidemic under 'Duration' definition", fontsize = 16)
plt.xlabel("Time (T days)", fontsize = 14)
plt.xscale('log')
plt.xlim(1,6000)
plt.ylabel("Probability persist for at least T days", fontsize = 14)
plt.ylim(0,1)
plt.legend(fontsize = 12)
```

## Extensions of the SIR model
A common way the real-world realism of the SIR model can be improved is by extending it to the SEIR model. The E stands for "Exposed" and is added to account for diseases which have a incubation period, meaning that individuals are infected, but not yet infectious. Notable diseases with an incubation period are COVID-19, measles and tuberculosis. This extension allows the model to more accurately predict the timing of outbreaks, allowing better timing of preventative measures like quarantines. Another similar way the model can be adapted to improve the prediction of outbreaks to include a waning period in which the immunity of recovered individuals diminishes over time. This reflects the reduction in the effectiveness of specific antibodies in the recovered individuals immune system. This is relevant for diseases like COVID-19 which has several strands which, due to mutations, cannot be combatted by the same antibodies as the original disease. This can be used to better time booster shots for vaccinations.

A different way the accuracy of the model can be improved is by splitting the three SIR compartments down into further strata for different characteristics of the population. One example is to incorporate age-structured compartments which allows the SIR model to better predict the outbreak of diseases such as influenza that disproportionately affect different age groups. Geographical location is another property which influences infection rates, specifically when the climate varies over a region which is important for diseases like malaria. By splitting the compartments up, different groups can have different infection rates and epidemiologists can investigate how a disease affects these different groups. This makes more targeted interventions possible, such as prioritising vaccinations for high-risk groups or focusing mosquito control efforts in specific regions.
